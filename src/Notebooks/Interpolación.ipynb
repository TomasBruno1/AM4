# Interpolación

Genero puntos aleatorios y los dibujo

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline, interp1d
# Genero puntos aleatorios
x = np.linspace(-3, 3, 50)
y = np.exp(-x**2) + 0.1 * np.random.randn(50)
# Grafico los puntos
plt.plot(x, y, 'ro', ms=5)
# Muestro el gráfico
plt.show()
```

Interpolación lineal

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline, interp1d
# Grafico los puntos
plt.plot(x, y, 'ro', ms=5)
linear = interp1d(x, y, kind='linear')
xs = np.linspace(-3, 3, 1000)
plt.plot(xs, linear(xs), 'g', lw=3)
# Muestro el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline, interp1d
# Grafico los puntos
plt.plot(x, y, 'ro', ms=5)
quad = interp1d(x, y, kind='quadratic')
xs = np.linspace(-3, 3, 1000)
plt.plot(xs, quad(xs), 'g', lw=3)
# Muestro el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline, interp1d
# Grafico los puntos
plt.plot(x, y, 'ro', ms=5)
cubic = interp1d(x, y, kind='cubic')
xs = np.linspace(-3, 3, 1000)
plt.plot(xs, cubic(xs), 'g', lw=3)
# Muestro el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline, interp1d
plt.plot(x, y, 'ro', ms=5)
spl = UnivariateSpline(x, y)
xs = np.linspace(-3, 3, 1000)
plt.plot(xs, spl(xs), 'g', lw=3)
# Muestro el gráfico
plt.show()
```

**Interpolación con splines por tramos**

```python
import matplotlib.pyplot as plt
from scipy.interpolate import InterpolatedUnivariateSpline
# Grafico los puntos
plt.plot(x, y, 'ro', ms=5)
# Interpolo con spline cúbico
spl = InterpolatedUnivariateSpline(x, y)
# Grafico el spline
xs = np.linspace(-3, 3, 1000)
plt.plot(xs, spl(xs), 'g', lw=3, alpha=0.7)
# Muestro el gráfico
plt.show()
```

**Cambio la suvaidad del spline**

```python
spl.set_smoothing_factor(0.10)
plt.plot(xs, spl(xs), 'b', lw=3)
plt.show()
```

**Dibujo del pato**

```python
from scipy.interpolate import UnivariateSpline
import matplotlib.pyplot as plt
import numpy as np

x = [0.9, 1.3, 1.9, 2.1, 2.6, 3.0, 3.9, 4.4, 4.7, 5.0, 6.0, 7.0, 8.0, 9.2, 10.5, 11.3, 11.6, 12.0, 12.6, 13.0, 13.3]
y = [1.3, 1.5, 1.85, 2.1, 2.6, 2.7, 2.4, 2.15, 2.05, 2.1, 2.25, 2.3, 2.25, 1.95, 1.4, 0.9, 0.7, 0.6, 0.5, 0.4, 0.25]

x1 = [0.9, 1.3, 1.9, 2.1, 2.6, 3.0, 3.9, 4.4, 4.7, 5.0, 6.0, 6.5, 7.0, 8.0, 9.2, 10.5, 11.3, 11.6, 12.0, 12.6, 13.0, 13.3]
y1 = [1.3, 1.5, 1.85, 2.1, 2.6, 2.7, 2.4, 2.15, 2.05, 2.1, 2.25, 4.00, 2.3, 2.25, 1.95, 1.4, 0.9, 0.7, 0.6, 0.5, 0.4, 0.25]

plt.plot(x, y, 'ro', ms=5)
plt.plot(x, y, 'go', ms=5)

spl = UnivariateSpline(x, y, s=0.000000001)
spl2 = UnivariateSpline(x1, y1, s=0.000000001)
xs = np.linspace(start=0.9, stop=13.3, num=10000)
plt.ylim([0, 15])
plt.xlim([0, 15])
plt.plot(xs, spl(xs), label='duck')
plt.plot(xs, spl2(xs), label='duck2')
plt.show()
```

**Comparando los distintos grados de splines**

```python
from scipy.interpolate import interp1d, UnivariateSpline
import matplotlib.pyplot as plt
import numpy as np

x = [0.9, 1.3, 1.9, 2.1, 2.6, 3.0, 3.9, 4.4, 4.7, 5.0, 6.0, 7.0, 8.0, 9.2, 10.5, 11.3, 11.6, 12.0, 12.6, 13.0, 13.3]
y = [1.3, 1.5, 1.85, 2.1, 2.6, 2.7, 2.4, 2.15, 2.05, 2.1, 2.25, 2.3, 2.25, 1.95, 1.4, 0.9, 0.7, 0.6, 0.5, 0.4, 0.25]

spl = interp1d(x, y, kind='quadratic')
spll = interp1d(x, y, kind='linear')
splll = interp1d(x, y, kind='cubic')
spl2 = UnivariateSpline(x, y)#, s=0.000000001)
xs = np.linspace(start=0.9, stop=13.3, num=10000)
plt.ylim([0, 15])
plt.xlim([0, 15])
plt.plot(xs, spl(xs), label='quadratic duck')
plt.plot(xs, spll(xs), label='linear duck')
plt.plot(xs, splll(xs), label='cubic duck')
plt.plot(xs, spl2(xs), label='univariate duck')
plt.legend()
plt.show()
```
