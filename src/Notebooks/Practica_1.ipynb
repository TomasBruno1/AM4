```python
import math

def raices_cuadratica(a, b, c):
    x1 = (-b + math.sqrt(b ** 2 - 4 * a * c)) / (2.0 * a)
    x2 = (-b - math.sqrt(b ** 2 - 4 * a * c)) / (2.0 * a)
    return x1, x2

def raices_cuadratica_mejorada(a, b, c):
    x2 = (-b - math.sqrt(b ** 2 - 4 * a * c)) / (2.0 * a)
    x1 = c / (a * x2)
    return x1, x2
```

```python
raices_cuadratica(1, 62.10, 1)
```

```python
raices_cuadratica_mejorada(1, 62.10,1)
```

```python
for x in range(1, 10):
    b = 10 ** x
    print("%15s | RC: %30s | RCM: %30s" % (b, raices_cuadratica(1,b,1), raices_cuadratica_mejorada(1,b,1)))
```
